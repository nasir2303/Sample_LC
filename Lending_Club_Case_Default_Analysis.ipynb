Removal of outliers from using IQR method

```python
Q1 = df["annual_inc"].quantile(0.25)
Q3 = df["annual_inc"].quantile(0.75)
IQR = Q3 - Q1

df = df[~((df["annual_inc"] < (Q1 - 1.5 * IQR)) | (df["annual_inc"] > (Q3 + 1.5 * IQR)))]
```

```python
plt.boxplot(df.annual_inc)
plt.show()
```

As there few more outliers present in the column and needs to be removed .
